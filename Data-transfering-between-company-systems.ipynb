```python
def clientOrigin(name):
    
    driverOrigen.get("https://business.solarmarket.com.br/clientes")
    try:
        nomear = WebDriverWait(driverOrigen, 30).until(EC.presence_of_element_located((By.XPATH,'/html/body/div/div[2]/div[1]/main/div/div/div/div[2]/div/div/input')))
    except:
        print("ERRO")
    else:
        nomear.send_keys(name)
        time.sleep(3)
        try:
            elemento = WebDriverWait(driverOrigen, 5).until(EC.presence_of_element_located((By.TAG_NAME, 'table')))
            
        except:
            print("\nCliente:" + name + " NAO Encontrado")
            return (0)
        element = elemento.find_elements(By.TAG_NAME, 'a')
        for e in element:
            #print(e.text)
            if e.text.lower() == name.lower() or len(element) == 1:
                print("\nCliente:" + name + " Encontrado")
                print("\nNumero de nomes Similares = ", len(element), "\n")
                ActionChains(driverOrigen).move_to_element(e).click(e).perform()
                return (1)

        print("\nCliente:" + name + " NAO Encontrado")
        return (0)
```
